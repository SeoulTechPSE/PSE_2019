## 1. 버전 관리란?


* 버전 관리 시스템(VCS - Version Control System)
 * 파일 변화를 시간에 따라 기록했다가 나중에 특정 시점의 버전을 다시 꺼내올 수 있는 시스템
 * 소프트웨어 소스 코드 뿐만 아니라 컴퓨터 파일의 버전도 관리할 수 있음
 * 이미지나 레이아웃도 적용 가능(for 그래픽, 웹 디자이너)
 * 각 파일 또는 프로젝트 전체를 이전 상태로 되돌릴 수 있음
 * 시간에 따라 수정 내용을 비교, 문제를 추적, 프로젝트의 기원을 알 수 있음
 * 파일을 잃어버리거나 잘못 고쳤을 때도 쉽게 복구할 수 있음

<br>
* 로컬 버전 관리(Local VCS, RCS - Revision Control System)
 * 파일 버전을 관리하기 위해 파일의 사본을 만들다 보면 실수하기가 쉬움
 * 이런 실수를 방지하기 위해 로컬 VCS가 탄생함
 * RCS는 Patch Set(파일에서 변경되는 부분)을 관리함
 * 일련의 Patch Set을 적용해서 모든 파일을 특정 시점으로 되돌릴 수 있음
 
![로컬 버전 관리](fig1-1.png)

<br>
* 중앙집중식 버전 관리(CVCS - Central VCS)
 * 다른 개발자와 함께 작업해야 하는 경우를 위해 중앙집중식 VCS가 개발됨
 * 서버가 별도로 있고 클라이언트가 중앙 서버에서 파일을 받아서 사용(Checkout)함
 * CVS, Subversion, Perforce 등이 있음
 * 모두 누가 무엇을 하고 있는지 알 수 있음
 * 관리자는 무엇을 할지 꼼꼼하게 관리할 수 있으며, 여러 로컬 DB를 관리하는 것보다 쉬움
 * 중앙 서버에 문제가 발생하면 아무것도 할 수 없음
 * 중앙 DB가 있는 하드디스크에 문제가 생기면 프로젝트의 모든 히스토리를 잃음

![중앙집중식 버전 관리](fig1-2.png)

<br>
* 분산 버전 관리 시스템(Distributed VCS)
 * 저장소를 통째로 복제하여 클라이언트 컴퓨터에서 작업함
 * 단순히 파일의 마지막 스냅샷만 Checkout하지 않고, 모든 히스토리를 복제함
 * 서버에 문제가 생기거나, 인터넷 연결이 없어도 작업할 수 있음
 * 여러 리모트 저장소를 통해 다른 사람들과 협업할 수 있음(워크플로가 다양함)
 * 리모트 저장소: 인터넷이나 네트워크 어딘가에 있는 저장소
 * Git, Mecurial, Bazaar, Darcs 등이 있음

![분산 버전 관리](fig1-3.png)

## 2. 짧게 보는 Git의 역사

* 규모가 큰 오픈소스 프로젝트인 Linux 커널에서 파일 관리는 Patch와 단순 압축 파일로 하였음 (1991-2002)
* 2002년에 Linux 커널은 상용 DVCS인 BitKeeper를 사용함
* Linux 커널은 오픈소스로 비영리적 <-> BitKeeper는 영리적 (2005)
* Linux 개발 커뮤니티가 자체 도구인 Git을 만드는 계기가 되었음
* 목표: 빠른 속도, 단순한 구조, 동시 다발적 가지 구조의 개발, 완벽한 분산, 대형 프로젝트에 적합(속도, 데이터 크기 등)
* Git은 속도가 빨라서 대형 프로젝트에 좋고, 동시다발적 브랜치를 다룰 수 있는 슈퍼 울트라 브랜칭 시스템임
<br>

## 3. Git 기초

* 델타 기반 버전 관리 시스템
 * 파일들의 목록을 관리함
 * 각 파일의 변화를 시간순으로 관리하면서 파일들의 집합을 관리함
 * CVS, Subversion, Perforce, Bazaar 등

![델타 기반 버전 관리](fig1-4.png)
<center>각 파일에 대한 변화를 저장하는 시스템</center>

<br>
* Git
 * 데이터를 파일 시스템 스냅샷의 연속으로 취급하고 크기가 아주 작음
 * 커밋하거나 프로젝트의 상태를 저장할 때마다 파일이 존재하는 순간을 중요하게 여김
 * 파일이 달라지지 않았으면 새로 저장하지 않고, 이전 상태 파일의 링크를 저장함
 * 데이터를 스냅샷의 스트림처럼 취급함
 * 3장 Git 브랜치에서 추가 설명

![git 버전 관리](fig1-5.png)
<center>시간순으로 프로젝트의 스냅샷을 저장</center>

<br>
* Git은 로컬에서 작업
 * 로컬 파일과 데이터만 사용하기 때문에 네트워크, 다른 컴퓨터가 필요 없음
 * CVCS는 네트워크 속도에 영향을 받지만 Git은 그렇지 않음
 * 프로젝트의 히스토리가 로컬 디스크에 있으므로 모든 명령이 순식간에 실행됨
 * 오프라인 상태이거나 VPN에 연결하지 못해도 작업할 수 있음

<br>
* Git의 무결성
 * 데이터를 저장하기 전에 항상 체크섬을 구하고 그것으로 데이터를 관리함
 * 체크섬: 가장 기본적인 데이터 단위이며 Git의 기본 철학
 * Git 없이는 체크섬을 다룰 수 없어서 파일의 상태를 알 수 없음
 * SHA-1 해시를 통하여 체크섬을 만들고 파일을 저장함
 * 체크섬은 `24b9da6552252987aa493b52f8696cd6d3b00373` 와 같은 40자 길이의 16진수 문자열

<br>
* 체크섬(checksum)
 * 중복검사의 한 형태로, 오류 정정을 통해 시공간 속에서 자료의 무결성을 보호하는 방법
 * 나열된 데이터를 더하여 체크섬 숫자를 얻고, 정해진 비트수로 재구성함
 * 체크섬이 맞아 떨어지는지 확인하여 데이터의 손상 여부를 확인할 수 있음
 * 체크섬의 예
 
> 4 바이트의 데이터: 0x25, 0x62, 0x3F, 0x52
<br> 1단계: 모든 바이트를 덧셈하면 0x118
<br> 2단계: 캐리 니블(0.5바이트=4비트=16진수=hex digit) 버림 0x18
<br> 3단계: 0x18의 2의 보수인 0xE8, 체크섬 바이트
<br> 확인: 원래 그룹의 바이트에 체크섬 바이트까지 모두 더하면 0x200이 됨
<br>　 　 다시 캐리 니블을 버리면 0x00이 됨, 오류가 없다는 뜻(오류가 있어도 우연히 0x00일 수도 있음)

 https://ko.wikipedia.org/wiki/%EC%B2%B4%ED%81%AC%EC%84%AC [체크섬]
 
 https://ko.wikipedia.org/wiki/2%EC%9D%98_%EB%B3%B4%EC%88%98 [2의 보수]

<br>

* 데이터의 추가
 * Git으로 데이터를 추가만 될 뿐 되돌리거나 삭제되지 않음
 * 스냅샷을 커밋하고 나면 데이터를 잃어버리기는 어려움
 * 되돌리기(2장)가 쉬우므로 즐겁게 여러 가지 실험을 할 수 있음

<br>
* 파일의 세 가지 상태
 * Committed: 데이터가 로컬 데이터베이스에 안전하게 저장됨
 * Modified: 수정한 파일을 아직 커밋하지 않은 상태
 * Staged: 현재 수정한 파일을 곧 커밋할 것이라고 표시한 상태
<br>
![파일의 세 가지 상태](fig1-6.png)
<center>워킹 트리, Staging Area, Git 디렉토리</center>
<br>

* Git 디렉토리
 * Git이 프로젝트의 메타데이터와 객체 데이터베이스를 저장하는 곳
 * 다른 컴퓨터에 있는 저장소를 Clone 할 때 Git 디렉토리가 만들어짐


* 워킹트리(Working Directory)
 * 프로젝트의 특정 버전을 Checkout 한 것임
 * 로컬 디스크에 있는 Clone 한 Git 디렉토리의 데이터베이스에서 파일을 가져와 워킹트리를 만듦


* Staging Area(Index)
 * Git 디렉토리에 있음
 * 곧 커밋할 파일에 대한 정보를 저장함


* Git에서 하는 일
 1. 워킹 트리에서 파일을 수정한다.
 2. Staging Area에 파일을 Stage해서 커밋할 스냅샷을 만든다. (일부 처리도 가능)
 3. 파일들을 커밋해서 Git 디렉토리에 영구적인 스냅샷으로 저장한다.
 * Git 디렉토리에 있는 파일들은 Committed 상태
 * 파일을 수정하고 Staging Area에 추가했다면 Staged 상태
 * Staging Area에 추가하지 않았으면 Modified 상태

## 4. CLI

## 6. Git 최초 설정

* 환경 설정
 * Git의 설치 후 사용 환경을 적절하게 설정해 주어야 함
 * `git config` 도구로 설정


1. `/etc/gitconfig` 파일: 시스템의 모든 사용자, 저장소에 적용
<br> `git config --system` 옵션으로 수정 가능
2. `~/.gitconfig`, `~/.config/git/config` 파일: 특정 사용자에게만 적용
<br> `git config --global` 옵션으로 수정 가능
3. `.git/config` 파일: Git 디렉토리의 특정 저장소(or 작업 중인 프로젝트)에만 적용
<br> `--local` 옵션으로 수정 가능(기본적으로 적용되어 있음)

 * 각 설정은 역순으로 우선시 된다. 3 > 2 > 1
 * Windows 에서는 `$HOME` 디렉토리에서 `.gitconfig`파일을 찾는다. (주로 `C:\Users\$User`)
 * `/etc/gitconfig` 파일은 그 경로(MSys 루트의 상대경로)에서 찾는다.
 * MSys 루트는 인스톨러로 Git을 Windows에 설치할 때 결정된다.
 * Git for Windows 2 버전에서는 `C:\ProgramData\Git\config` (Vista 이상)
 * 시스템 설정 파일의 경로는 `git config -f <file>` 명령으로 변경할 수 있다.
 

* 사용자 정보
 * Git은 커밋할 때마다 사용자이름과 이메일 주소를 사용함
 * 한 번 커밋한 후에는 정보를 변경할 수 없음 <br>
`$ git config --global user.name "John Doe"` <br>
`$ git config --global user.email johndoe@example.com` (github에 가입한 이메일)
 * 프로젝트마다 다른 이름, 이메일을 사용하고 싶으면 `--global` 옵션을 빼면 됨
 * `--global` 옵션 설정은 딱 한 번만 하면 됨
![git config](fig1-11.png)
<center>git config 사용자 정보</center>
<br>


* 편집기
 * 사용자 정보를 설정한 뒤 Git에서 사용할 텍스트 편집기 설정
 * 기본적으로 Git은 시스템의 기본편집기를 사용함
 * Emacs 같은 다른 편집기를 사용할 때는 
<br>`$ git config --global core.deitor emacs`
 * Windows 에서는 프로그램의 경로를 입력하면 사용 가능 (ex. notepad)
<br>`$ git config --global core.editor "'C:/Program Files/Notepad++/notepad++.exe' -multiInst -nosession"`
<br>

* 설정 확인
 * `git config --list`을 실행하면 설정한 것들을 보여줌
 * `git config <key>`을 실행하면 특정 Key의 값을 알 수 있음

## 7. 주요 명령어


* git init: 깃 저장소를 초기화한다. 저장소나 디렉토리 안에서 이 명령을 실행하기 전까지는 그냥 일반 폴더이다. 이것을 입력한 후에야 추가적인 깃 명령어들을 줄 수 있다.


* git config: “configure”의 준말, 처음에 깃을 설정할 때 가장 유용하다.


* git help: 명령어를 잊어버렸다? 커맨드 라인에 이걸 타이핑하면 21개의 가장 많이 사용하는 깃 명령어들이 나타난다. 좀 더 자세하게 “git help init”이나 다른 용어를 타이핑하여 특정 깃 명령어를 사용하고 설정하는 법을 이해할 수도 있다.


* git status: 저장소 상태를 체크. 어떤 파일이 저장소 안에 있는지, 커밋이 필요한 변경사항이 있는지, 현재 저장소의 어떤 브랜치에서 작업하고 있는지 등을 볼 수 있다.


* git add: 이 명령이 저장소에 새 파일들을 추가하진 않는다. 대신, 깃이 새 파일들을 지켜보게 한다. 화일을 추가하면, 깃의 저장소 “스냅샷”에 포함된다.


* git commit: 깃의 가장 중요한 명령어. 어떤 변경사항이라도 만든 후, 저장소의 “스냅샷”을 찍기 위해 이것을 입력한다. 보통 “git commit -m “Message hear.” 형식으로 사용한다. `-m`은 명령어의 그 다음 부분을 메시지로 읽어야 한다는 것을 말한다.


* git branch: 여러 협업자와 작업하고 자신만의 변경을 원한다? 이 명령어는 새로운 브랜치를 만들고, 자신만의 변경사항과 화일 추가 등의 커밋 타임라인을 만든다. 당신의 제목이 명령어 다음에 온다. 새 브랜치를 “cats”로 부르고 싶으면, `git branch cats`를 타이핑한다.


* git checkout: 글자 그대로, 현재 위치하고 있지 않은 저장소를 “체크아웃”할 수 있다. 이것은 체크하길 원하는 저장소로 옮겨가게 해주는 탐색 명령이다. master 브랜치를 들여다 보고 싶으면, `git checkout master`를 사용할 수 있고, `git checkout cats`로 또 다른 브랜치를 들여다 볼 수 있다.


* git merge: 브랜치에서 작업을 끝내고, 모든 협업자가 볼 수 있는 master 브랜치로 병합할 수 있다. `git merge cats`는 “cats” 브랜치에서 만든 모든 변경사항을 master로 추가한다.


* git push: 로컬 컴퓨터에서 작업하고 당신의 커밋을 깃허브에서 온라인으로도 볼 수 있기를 원한다면, 이 명령어로 깃허브에 변경사항을 “push”한다.


* git pull: 로컬 컴퓨터에서 작업할 때, 작업하고 있는 저장소의 최신 버전을 원하면, 이 명령어로 깃허브로부터 변경사항을 다운로드한다(“pull”).

## 9. 기본 용어


* 커맨드 라인(Command Line): 깃 명령어를 입력할 때 사용하는 컴퓨터 프로그램. 맥에선 터미널이라고 한다. PC에선 기본적인 프로그램이 아니어서 처음엔 깃을 다운로드해야 한다. 두 경우 모두 마우스를 사용하는 것이 아닌 프롬프트로 알려진 텍스트 기반 명령어를 입력한다.


* 저장소(Repository): 프로젝트가 거주(live)할 수 있는 디렉토리나 저장 공간. 깃허브 사용자는 종종 “repo”로 줄여서 사용한다. 당신의 컴퓨터 안의 로컬 폴더가 될 수도 있고, 깃허브나 다른 온라인 호스트의 저장 공간이 될 수도 있다. 저장소 안에 코드 화일, 텍스트 화일, 이미지 화일을 저장하고, 이름붙일 수 있다.


* 버전관리(Version Control): 기본적으로, 깃이 서비스되도록 고안된 목적. MS 워드 작업할 때, 저장하면 이전 화일 위에 겹쳐쓰거나 여러 버전으로 나누어 저장한다. 깃을 사용하면 그럴 필요가 없다. 프로젝트 히스토리의 모든 시점의 “스냅샷”을 유지하므로, 결코 잃어버리거나 겹쳐쓰지 않을 수 있다.


* 커밋(Commit): 깃에게 파워를 주는 명령이다. 커밋하면, 그 시점의 당신의 저장소의 “스냅샷”을 찍어, 프로젝트를 이전의 어떠한 상태로든 재평가하거나 복원할 수 있는 체크포인트를 가질 수 있다.


* 브랜치(Branch): 여러 명이 하나의 프로젝트에서 깃 없이 작업하는 것이 얼마나 혼란스러울 것인가? 일반적으로, 작업자들은 메인 프로젝트의 브랜치를 따와서(branch off), 자신이 변경하고 싶은 자신만의 버전을 만든다. 작업을 끝낸 후, 프로젝트의 메인 디렉토리인 “master”에 브랜치를 다시 “Merge”한다.
